```python
from coffragmentorllm.pdf import COFPaper

OPENAI_KEY = "YOUR OPENAI API KEY"
```

### COF Paper

```python
FILE1 = 'pdf/101002adma200800030.pdf'
FILE2 = 'pdf/10.1038_nchem.695_1.pdf'
FILE3 = 'pdf/10.1038_s41557-018-0141-5_0.pdf'

file = COFPaper(FILE1, verbose=True)

print(f'Lenght of paper: {len(file.content)}')
display(file.content[:100])
```

### Extract linkers of COFs (fragment) with gpt-4o (OpenAI)

```python
fragments = file.fragment(openai_key=OPENAI_KEY)
```

### Analyze extraction

```python
def summary_paper(fragments):
    print(f"n COFs detected in paper: {len(fragments)}")
    
    for cof_fragments in fragments:
        print('\n---------')
        print(f'COF name: {cof_fragments.name}')
        print(f'COF fragments (as extracted): {cof_fragments.linkers}')
        print(f'COF fragments abbreviations (as extracted): {cof_fragments.linkers_abbr}')
        print(f'COF linkage type (as extracted): {cof_fragments.linkage}')

        print(f'COF fragments search in PubChem: {[linker.pubchem for linker in cof_fragments.linkers]}')
        print(f'COF fragments\' IUPAC names (as PubChem): {[linker.iupac for linker in cof_fragments.linkers]}')
        cof_fragments.linkers.display(figsize = (5,5))



summary_paper(fragments)
```
